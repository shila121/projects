```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score

%matplotlib inline
```

```python
# read the df
df =  pd.read_csv('bankloans.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.value_counts()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
fig,ax =  plt.subplots(figsize = (10,10))
sns.lineplot(x = 'age' ,y = 'income', data = df,ax=ax)
```

```python
fig,ax = plt.subplots(figsize=(20,10))
sns.lineplot(x='age',y='debtinc',data=df,ax=ax)
```

```python
df['default'].value_counts()
```

```python
x =  df.drop(['default'],axis = 1)
y =  df['default']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
sc = StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.fit_transform(xtest)
```

```python
# Creating Model

# Random Forest


```

```python
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(xtrain,ytrain)
```

```python
rfc.score(xtest,ytest)
```

```python
rfc2 = cross_val_score(estimator=rfc,X=xtrain,y=ytrain,cv=10)
rfc2.mean()
```

```python
# SVM


```

```python
sv = SVC()
sv.fit(xtrain,ytrain)
```

```python
sv.score(xtest,ytest)
```

```python
model = GridSearchCV(sv,{
    'C':[0.1,0.2,0.4,0.8,1.2,1.8,4.0,7.0],
    'gamma':[0.1,0.4,0.8,1.0,2.0,3.0],
    'kernel':['rbf','linear']
},scoring='accuracy',cv=10)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.best_params_
```

```python
model2 = SVC(C=0.1,gamma=0.1,kernel='linear')
model2.fit(xtrain,ytrain)
model2.score(xtest,ytest)
```

```python
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
lr.score(xtest,ytest)
```

```python
yp = lr.predict(xtest)
c= confusion_matrix(ytest,yp)
fig ,ax = plt.subplots(figsize=(5,5))
sns.heatmap(c,ax=ax)
```
